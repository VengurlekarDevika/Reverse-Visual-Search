```python
import numpy as np
import pandas as pd
import random
import os
import shutil
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
```

```python
file_path= r"C:\\Users\\devik\\Downloads\\coco_minitrain_25k\\coco_minitrain_25k"
```

```python
labels_categories=os.listdir(file_path+"\\labels\\train2017")
```

```python
person_images=[]
for i in labels_categories:
    if list(pd.read_csv(file_path+"\\labels\\train2017\\"+i,delimiter=" ",header=None)[0]).count(0)>0:
        person_images.append(i)
```

```python
print(person_images)
```

```python
images_path="C:\\Users\\devik\\Downloads\\coco_minitrain_25k\\coco_minitrain_25k\\images\\train2017\\"

for v in random.sample(range(0,13629),10):
    images = mpimg.imread(images_path+person_images[v].split(".")[0]+".jpg")
    plt.imshow(images)
    plt.show()
    
```
